```python
# Doubly Circular LL
class Node:
    def __init__(self, elem, next = None, prev = None):
        self.elem = elem
        self.next = next
        self.prev = prev
```

```python
def create_DLL(arr):
    dh = Node(None)
    dh.next = dh
    dh.prev = dh
    for i in range(len(arr)):
        new_node = Node(arr[i])
        dh.prev.next = new_node
        new_node.prev = dh.prev
        new_node.next = dh
        dh.prev = new_node
    return dh
```

```python
import numpy as np
head = create_DLL(np.array([5, 10, 15, 20, 30]))
```

```python
def print_LL(dh):
    temp = dh.next
    while temp.elem != None:
        print(temp.elem, end = " -> ")
        temp = temp.next
    return
```

```python
print_LL(head)
```

```python
# Approach 1
def reverse_DLL(dh):
    dh1 = Node(None)
    dh1.next = dh1
    dh1.prev = dh1

    temp = dh.next
    while temp.elem != None:
        new_node = Node(temp.elem)
        new_node.next = dh1.next
        new_node.prev = dh1
        dh1.next.prev = new_node
        dh1.next = new_node
        temp = temp.next
    return dh1

def length(dh):
    count = 1
    temp = dh.next
    while temp.elem != None:
        count += 1
        temp = temp.next
    return count


def reverse_DLL2(dh):
    size = length(dh)
    a = dh.next
    b = dh.prev
    for i in range(size//2):
        a.elem, b.elem = b.elem, a.elem
        a = a.next
        b = b.prev
    return dh

# driver code
head = create_DLL(np.array([5, 10, 15, 20, 30]))
head2 = reverse_DLL2(head)
print_LL(head2)
```
