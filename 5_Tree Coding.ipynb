```python
class BTNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
N60 = BTNode(60)
N50 = BTNode(50)
N20 = BTNode(20)
N10 = BTNode(10)
N5 = BTNode(5)
N6 = BTNode(6)
N30 = BTNode(30)
```

```python
N50.left = N30
N50.right = N60
N30.left = N20
N60.right = N10
N10.left = N5
N10.right = N6
```

```python
def post_Order(root):
    if root == None:
        return
    post_Order(root.left)
    post_Order(root.right)
    print(root.value, end = " ")
    return
```

```python
post_Order(N50)
```

```python
def createTreeFromArray(arr, i):
    if i >= len(arr) or arr[i] == None:
        return None
    root = BTNode(arr[i])
    root.left = createTreeFromArray(arr, 2*i)
    root.right = createTreeFromArray(arr, 2*i+1)
    return root
```

```python
import numpy as np
arr = np.array([None, 100, 50, 60, None, 40, 70, 10, None, None, 80, 90, None, 15, 5, 200])
```

```python
new_tree = createTreeFromArray(arr, 1)
post_Order(new_tree)
# 80 90 40 50 15 70 5 200 10 60 100
```

```python
# Sec 34
class BTNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
def postOrder(root):
    if root == None:
        return
    postOrder(root.left)
    postOrder(root.right)
    print(root.value, end = " ")
    return
```

```python
def preOrder(root):
    if root == None:
        return
    print(root.value, end = " ")
    preOrder(root.left)
    preOrder(root.right)
```

```python
N50 = BTNode(50)
N10 = BTNode(10)
N30 = BTNode(30)
N5 = BTNode(5)
N6 = BTNode(6)
N20 = BTNode(20)
```

```python
N50.left = N10
N50.right = N30
N10.left = N20
N30.left = N5
N30.right = N6
```

```python
# 20 10 5 6 30 50
preOrder(N50)
```

```python
def createTreeFromArray(arr, index_or_root):
    #Base case
    if index_or_root >= len(arr) or arr[index_or_root] == None:
        return None
    root = BTNode(arr[index_or_root])
    root.left = createTreeFromArray(arr, 2*index_or_root)
    root.right = createTreeFromArray(arr, 2*index_or_root+1)
    return root
```

```python
import numpy as np
arr = np.array([None, 50, 10, 30, 20, None, 5, 6, None, 70, None, None, 25, None, 60, 90])
```

```python
print(arr)
```

```python
Tree = createTreeFromArray(arr, 1)
postOrder(Tree)
# 70 20 10
```

```python
def createArrayFromTree(root, arr, index):
    if root == None or index >= len(arr):
        return
    arr[index] = root.value
    createArrayFromTree(root.left, arr, 2*index)
    createArrayFromTree(root.right, arr, 2*index+1)
```
