```python
def insert(root, value):
    if root == None:
        node = BTNode(value)
        return node
    if root.value == value:
        print("Value already exists")
        return root
    if value > root.value:
        root.right = insert(root.right, value)
    else:
        root.left = insert(root.left, value)
    return root
```

```python
def successor(root):
    if root.left == None:
        return root.value
    else:
        return successor(root.left)
```

```python
def remove(root, value):
    # Base case
    if root == None:
        print("No value found!")
        return None

    if value > root.value:
        root.right = remove(root.right, value)
    elif value < root.value:
        root.left = remove(root.left, value)
    if value == root.value:
        # No child
        if root.left == None and root.right == None:
            return None
        # one child
        elif root.left == None:
            return root.right
        elif root.right == None:
            return root.left
        # two child
        else:
            successor_value = successor(root.right)
            root.value = successor_value
            root.right = remove(root.right, successor_value)
    return root

```
