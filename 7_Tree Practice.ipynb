```python
class BTNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def post_Order(root):
    if root == None:
        return
    post_Order(root.left)
    post_Order(root.right)
    print(root.value, end = " ")
    return

def in_Order(root):
    if root == None:
        return
    in_Order(root.left)
    print(root.value, end = " ")
    in_Order(root.right)
    return
```

```python

root = BTNode(50)
N60 = BTNode(60)
N20 = BTNode(20)
N10 = BTNode(10)
N5 = BTNode(5)
N6 = BTNode(6)
N30 = BTNode(30)

root.left = N30
root.right = N60
N30.left = N20
N60.right = N10
N10.left = N5
N10.right = N6
```

```python
in_Order(root)
```

```python
def left_Sum(root, child_type = "root"):
    if root == None:
        return 0
    total = 0
    if child_type == "left":
        total += root.value
    total += left_Sum(root.left, child_type = "left")
    total += left_Sum(root.right, child_type = "right")
    return total
```

```python
print(left_Sum(root, "root"))
```

```python
def print_path(root, search_value):
    if root == None:
        return "-1"
    if root.value == search_value:
        return str(root.value)
    left = print_path(root.left, search_value)
    right = print_path(root.right, search_value)
    if left == "-1":
        if right == "-1":
            return "-1"
        else:
            return str(root.value) + " -> " + right
    else:
        return str(root.value) + " -> " + left
```

```python
print(print_path(root, 5))
```
