```python
class Node:
    def __init__(self, elem, next = None):
        self.elem = elem
        self.next = next

a = Node(10)
b = Node(20)
c = Node(40)
a.next = b
b.next = c
c.next = None
```

```python
print(a)
```

```python
def print_LL(head):
    temp = head
    while temp != None:
        print(temp.elem, end = " ")
        temp = temp.next
```

```python
print_LL(a)
```

```python
def create_LL(arr):
    head = Node(arr[0])
    tail = head
    for i in range(1, len(arr)):
        new_node = Node(arr[i])
        tail.next = new_node
        tail = new_node
    return head
```

```python
# arr = [10, 20, 30, 40]
import numpy as np
arr =np.array([10, 20, 30, 40, 50, 60, 70])
x = create_LL(arr)
```

```python
print_LL(x)
```

```python
def insert(head, position, new_val):
    temp = head
    count = 0
    while temp!= None:
        if count == position:
            new_node = Node(new_val)
            new_node.next = temp.next
            temp.next = new_node
            return head
        count += 1
        temp = temp.next
```

```python
y = insert(x, 2, 100)
```

```python
print_LL(y)
```

```python
def remove(head, position):
    temp = head
    count = 0
    while temp!= None:
        if count == position-1:
            x = temp.next
            temp.next = temp.next.next
            x.next = None
            return head
        count += 1
        temp = temp.next
```

```python
z = remove(y, 2)
print_LL(z)
```

```python
def sort_LL(head):
    while True:
        temp = head
        flag = True
        while temp.next != None:
            if temp.elem > temp.next.elem:
                flag = False
                m = temp.elem
                temp.elem = temp.next.elem
                temp.next.elem = m
            temp = temp.next
        if flag == True:
            return head
```

```python
y = sort_LL(x)
print_LL(y)
```

```python
def concate(head1, head2, insert_elem):
    temp1 = head1
    temp2 = head2
    while True:
        if temp1.elem == insert_elem:
            break
        temp1 = temp1.next
    while temp2.next != None:
        temp2 = temp2.next
    temp2.next = temp1.next
    temp1.next = head2
    return head1
```

```python
print_LL(y)
print()
print_LL(z)
```

```python
x = concate(y, z, 30)
print_LL(x)
```

```python
# section 34
# Swap first and last element swap
def swap(head):
    temp = head
    while temp.next != None:
        temp = temp.next
    m = head.elem
    head.elem = temp.elem
    temp.elem = m
    return head
```

```python
y = swap(x)
print_LL(y)
```

```python
def remove_same(head, value):
    temp =  head
    count = 0
    if head.elem == value:
        count += 1
    while temp.next != None:
        if temp.next.elem == value:
            count += 1
            if count > 1:
                temp.next = temp.next.next
            else:
                temp = temp.next
        else:
            temp = temp.next
    return head
```

```python
arr = np.array([20, 10, 30, 40, 10,10, 60, 70])
x = create_LL(arr)
print_LL(x)
print()
x = remove_same(x, 10)
print_LL(x)
```

```python
def even_odd(head):
    new_node = Node(-1)
    new_node.next = head
    head = new_node
    temp = head
    head2 = None #odd LL
    tail2 = None
    odd_count = 0

    while temp.next != None:
        if temp.next.elem % 2 == 0: #even
            temp = temp.next
        else: #odd
            odd_count += 1
            if odd_count == 1:
                head2 = temp.next
                tail2 = head2
                temp.next = temp.next.next
            else:
                tail2.next = temp.next
                tail2 = temp.next
                temp.next = temp.next.next
    temp = head
    while temp.next != None:
        temp = temp.next
    temp.next = head2
    return head.next
```

```python
arr = np.array([7, 1, 3, 4, 1,1, 6, 7])
x = create_LL(arr)
print_LL(x)
print("\nAfter change")
x = even_odd(x)
print_LL(x)
```
