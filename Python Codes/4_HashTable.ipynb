```python
from posixpath import sep
import numpy as np
class Node:
    def __init__(self, key, value, next = None):
            self.key = key
            self.value = value
            self.next = next

class HashTable:
    def __init__(self, size = 5):
        self.hashTable = np.array([None]*size)

    def hashFunction(self, key):
        f_c = key[0]
        l_c = key[len(key)-1]
        sum = ord(f_c) + ord(l_c)
        index = sum % len(self.hashTable)
        return index

    def insert(self, key, value):
        index = self.hashFunction(key)
        if self.hashTable[index] == None: # first insert
            new_node = Node(key, value)
            self.hashTable[index] = new_node
            print("Key and value has been inserted")
        else: # already there is a Linked List
            head = self.hashTable[index]
            new_node = Node(key, value)
            new_node.next = head
            head = new_node
            self.hashTable[index] = head
            print("Key and value has been inserted")
    def printLL(self,head):
        temp = head
        while temp != None:
            print(f"({temp.key}, {temp.value})", end = " -> ")
            temp = temp.next
    def printHashTable(self):
        for i in range(len(self.hashTable)):
            print(f"index {i}: ", end = " ")
            if self.hashTable[i] == None:
                continue
            else:
                # print LL
                self.printLL(self.hashTable[i])
            print()
```
