```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
    # if self.__top == None:
    #     return True
    # else:
    #     return False
```

```python
def print_stack(st):
  if st.isEmpty():
    return
  p = st.pop()
  print('|',p,end=' ')
  if p<10:
    print(' |')
  else:
    print('|')
  #print('------')
  print_stack(st)
  st.push(p)
```

```python
st = Stack()
```

```python
import numpy as np
arr = np.array([6, 9, 10, 25, 15, 5])
for i in range(len(arr)):
    st.push(arr[i])
```

```python
print_stack(st)
```

```python
def remove_max(st):
    s2 = Stack()
    max = st.peek()

    while st.isEmpty() != True:
        y = st.pop()
        if y > max:
            max = y
        s2.push(y)
    while s2.isEmpty() != True:
         y = s2.pop()
         if y != max:
            st.push(y)
    return st
```

```python
x = remove_max(st)
```

```python
print_stack(x)
```

```python
def remove_odd(st):
    if st.isEmpty() == True:
        return st
    else:
        new_stack = Stack()
        while st.isEmpty() != True:
            y = st.pop()
            if y % 2 == 0:
                new_stack.push(y)
        while new_stack.isEmpty() != True:
            # y = new_stack.pop()
            # st.push(y)
            st.push(new_stack.pop())
        return st
```

```python
x = remove_odd(st)
```
