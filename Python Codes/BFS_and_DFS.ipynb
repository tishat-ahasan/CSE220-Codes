```python
class Node:
    def __init__(self, elem):
        self.elem = elem
        self.next = None

class Queue:
    def __init__(self):
        self.front = self.rear = None

    def enqueue(self, elem):
        new_node = Node(elem)
        if self.rear is None:
            self.front = self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node

    def dequeue(self):
        if self.is_empty():
            raise IndexError("Queue is empty")
        removed_elem = self.front.elem
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return removed_elem

    def peek(self):
        if self.is_empty():
            raise IndexError("Queue is empty")
        return self.front.elem

    def is_empty(self):
        return self.front is None
```

```python
class Stack:
    def __init__(self):
        self.top = None

    def push(self, elem):
        new_node = Node(elem)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        popped_elem = self.top.elem
        self.top = self.top.next
        return popped_elem

    def peek(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        return self.top.elem

    def is_empty(self):
        return self.top is None
```

```python
import numpy as np
def BFS(graph, start):
    total_vertex = len(graph)
    visited = np.array([False]*total_vertex)
    Q = Queue()
    visited[start] = True
    Q.enqueue(start)
    while Q.is_empty() == False:
        u = Q.dequeue()
        print(u)
        for v in range(total_vertex):
            if graph[u][v] != 0 and visited[v] == False:
                visited[v] = True
                Q.enqueue(v)
```

```python
import numpy as np
def DFS(graph, start):
    total_vertex = len(graph)
    visited = np.array([False]*total_vertex)
    stk = Stack()
    visited[start] = True
    stk.push(start)
    while stk.is_empty() == False:
        u = stk.pop()
        print(u)
        for v in range(total_vertex):
            if graph[u][v] != 0 and visited[v] == False:
                visited[v] = True
                stk.push(v)
```
