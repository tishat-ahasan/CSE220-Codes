```python
def leftShift(arr):
    temp = arr[0]
    for i in range(len(arr)-1):
        arr[i] = arr[i+1]
        print("for i = ", i, "array = ", arr)
    arr[len(arr)-1] = temp
    return arr
```

```python
print(arr)
leftShift(arr)
```
